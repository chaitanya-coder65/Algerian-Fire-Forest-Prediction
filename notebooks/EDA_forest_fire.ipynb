```python
import pandas as pd 
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
dataset.info()
```

## Data cleaning

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
dataset.head()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
## Changes required columns as integer datatypes 
```

```python
df[["month","day","year","Temperature","RH","Ws"]] = df[["month","day","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
## Changing other columns to float datatype 
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!="Classes":
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
## Saved the cleaned dataset
```

```python
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv",index=False)
```

## Exploretory Data Analysis

```python
df_copy = df.drop(['day','month','year'],axis = 1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encodeing of catogeris in classes

df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire',case=False),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## plot density plot all features
sns.set() 
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Pefcentage for pie chart
percentages = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
##Plotting pie chart
classlables = ["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentages,labels=classlables,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
